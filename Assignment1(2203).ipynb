*****Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the distribution.*****

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], bins=20, kde=True, color='skyblue', edgecolor='black')
plt.title('Distribution of Flight Prices')
plt.xlabel('Flight Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

*****Q3. What is the range of prices in the dataset? What is the minimum and maximum price?*****

```python
# Extract the 'Price' column
prices = df['Price']

# Calculate the minimum and maximum prices
min_price = prices.min()
max_price = prices.max()

# Calculate the range of prices
price_range = max_price - min_price

print("Minimum Price:", min_price)
print("Maximum Price:", max_price)
print("Price Range:", price_range)
```

*****Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize the relationship.*****

```python
#Dealing with 'Size' column
df['Size'].unique()

def convert_size(size):
  if 'M' in size:
    return int(float(size.replace("M", ''))*1000) #Here I used float() because some values are string value and multiplying string value with 1000 duplicates the same value 1000 times
  elif 'k' in size:
    return float(size.replace('k', ''))
  elif 'Varies with device' in size:
      return np.nan
  else:
    return np.nan

df['Size'] = df['Size'].apply(convert_size)

# Replacing null values with 0 in Rating column
df.fillna(0, inplace=True)
```

```python
# Scatter plot between size and rating of apps
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Size', y='Rating',color = 'blue', alpha=0.5)
plt.title('Relationship between App Size and Rating')
plt.xlabel('Size')
plt.ylabel('Rating')
plt.grid(True)
plt.show()
```

*****Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type*****

```python
not_numeric = df[~df['Price'].str.isnumeric()]
not_numeric
```

```python
df.loc[10472]
```

```python
# drop the entire row of index 10472
df.drop(index=10472, inplace=True)

# Remove '$' from the 'Price' column
df['Price'] = df['Price'].str.replace('$', '')

# Convert 'Price' column to numeric
df['Price'] = pd.to_numeric(df['Price'])

# Calculate average prices by app type
average_prices = df.groupby('Type')['Price'].mean()
average_prices
```

```python
# Creating bar chart to compare average prices by app type
plt.figure(figsize=(20, 10))
sns.barplot(x = average_prices.index, y  = average_prices.values, palette='viridis')
plt.xlabel('App_Type')
plt.ylabel('Average_Prices')
plt.xticks(rotation = 90)
plt.show()
```
